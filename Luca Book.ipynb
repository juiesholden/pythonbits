```python
import time

parkHours = int(input("How many hours do you want to park: "))

if parkHours <= 2:
    cost = 3.5
    seconds = 3600 * 2
elif parkHours <= 4:
    cost = 5.0
    seconds = 3600 * 4
elif parkHours <= 12:
    cost = 10.0
    seconds = 3600 * 12
else:
    print("Incorrect Duration")
    
currentTime = time.time()
expireTime = currentTime + seconds
print("Time now: ",time.ctime(currentTime))
print("Expires: ", time.ctime(expireTime))
print("Charge = ", cost)


```

```python
print("Code evaluator")

countAB = 0
count00 = 0

while True:
    code = input("Enter 5 digit product code: ")
    if len(code) !=5:
        print("Error: Wrong Length")
        break
        
    if code[0] == "A" and code[1] == "B":
        countAB += 1
    
    if code[3] == "0" and code[4] == "0":
        count00 += 1
    
    print(" AB codes", countAB)
    print(" 00 codes", count00)
```

```python
studentnames = ["A","B","C","D"]
results = []
topMark = 0
numberOfStudents = len(studentnames)

for student in range(numberOfStudents):
    print("Enter mark for", studentnames[student],": ", end= " ")
    mark = int(input())
    results.append(mark)
    if mark > topMark:
        topMark = mark
        topStudent = studentnames[student]

print("\nStudent \tExam mark")    
for student in range(numberOfStudents):
    print(studentnames[student],"\t\t",results[student])
    
print("\nTop result: ",topStudent,topMark)    
   

```

```python
name = ""
marks = []
mark = 0
totalMarks = 0

name = input("Please enter the student's name: ")
for i in range(0,5):
    print(" Please enter mark", i, "for", name, end = " ")
    mark = int(input())
    marks.append(mark)

for i in range(0,5):
    totalMarks += marks[i]
    
SortedMarks = sorted(marks, reverse = True) # sorted function
print("Results for", name)
print("Total is", totalMarks)
print("Topmarks are", SortedMarks[0:2])       # colon to separate indexes
print("Bottom marks are", SortedMarks[3:5])

```
